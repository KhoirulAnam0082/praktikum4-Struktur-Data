```python
# A Python Program to check if the given
# string is a pangram or not
def checkPangram(s):
 List = []
# create list of 26 characters and set false each entry
 for i in range(26):
  List.append(False)
 # converting the sentence to lowercase and iterating
 # over the sentence
 for c in s.lower():
  if not c == " ":
 # make the corresponding entry True
   List[ord(c) - ord('a')] = True
 # check if any character is missing then return False
 for ch in List:
   if ch == False:
    return False
    return True
# Driver Program to test above functions
sentence = "The quick brown fox jumps over the little lazy dog"
if (checkPangram(sentence)):
 print('"'+sentence+'"')
 print("\nis a pangram")
else:
 print('"'+sentence+'"')
 print("\nis not a pangram")
```

```python
import string

# Inisialisasi alfabet dalam bahasa Inggris
alfabet = string.ascii_lowercase

# Inisialisasi kalimat pangram kosong
kalimat_pangram = ""

# Buat kalimat pangram
for huruf in alfabet:
    if huruf not in kalimat_pangram:
        kalimat_pangram += huruf

# Cetak kalimat pangram yang telah dibuat
print("Kalimat Pangram yang Dibuat:", kalimat_pangram)
```

```python

def is_pangram(s):
    # Langkah 1: Buat vektor tanda (mark) boolean dengan ukuran 26
    mark = [False] * 26

    # Konversi semua huruf dalam string ke huruf kecil
    s = s.lower()

    # Langkah 2: Tandai huruf yang ada dalam string
    for char in s:
        if char.isalpha():
            # Tandai huruf yang valid
            mark[ord(char) - ord('a')] = True

    # Langkah 3: Periksa apakah semua indeks ditandai (semua karakter ada)
    return all(mark)

# Contoh penggunaan
input_string = "The quick brown fox jumps over the lazy dog"
result = is_pangram(input_string)

if result:
    print("Ini adalah pangram.")
else:
    print("Ini bukan pangram.")
```
